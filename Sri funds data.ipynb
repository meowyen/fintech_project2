```python
from pathlib import Path
import pandas as pd
```

```python
sri_funds_path = Path("SRI Funds - Sheet1.csv")
sri_funds_data = pd.read_csv(sri_funds_path)
sri_funds_data
# Key: X = No investment; P = Positive investment; R = Restricted investment; NS = No screens.
# defense/weapons animal testing, envionment, human rights
```

```python
weapons_defense_df = sri_funds_data.filter(items=['Fund','Defense/ Weapons'])
animal_testing_df = sri_funds_data.filter(items=['Fund','Animal Testing'])
envir_df = sri_funds_data.filter(items=['Fund','Environment'])
human_rights_df = sri_funds_data.filter(items=['Fund','Human Rights'])
```

```python
weapons_defense_df = weapons_defense_df.loc[weapons_defense_df['Defense/ Weapons'] =='X']
animal_testing_df = animal_testing_df.loc[animal_testing_df['Animal Testing'] == 'X']
envir_df = envir_df.loc[envir_df['Environment'] == 'P']
human_rights_df = human_rights_df.loc[human_rights_df['Human Rights'] == 'P']
```

```python
weapons_defense_list = weapons_defense_df['Fund'].tolist()
animal_testing_list = animal_testing_df['Fund'].tolist()
environment_list = envir_df['Fund'].tolist()
human_rights_list = human_rights_df['Fund'].tolist()
```

```python
funds_list = weapons_defense_list + animal_testing_list + environment_list + human_rights_list
funds_list
```
