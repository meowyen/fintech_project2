```python
import pandas as pd
from pathlib import Path
from dotenv import load_dotenv
import os
import re
import requests
import json

load_dotenv()
```

```python
global100_filepath = Path("data/global100_2021.csv")
global100_df = pd.read_csv(global100_filepath)

weapons_filepath = Path("data/weapons_output.csv")
weapons_df = pd.read_csv(weapons_filepath)


animal_filepath = Path("data/animal_output.csv")
animal_df = pd.read_csv(animal_filepath)

envir_filepath = Path("data/envir_output.csv")
envir_df = pd.read_csv(envir_filepath)

rights_filepath = Path("data/rights_output.csv")
rights_df = pd.read_csv(rights_filepath)

ctw_filepath = Path("data/change_the_world_2020.csv")
ctw_df = pd.read_csv(ctw_filepath)

```

```python
tradier_lookup_url = "https://sandbox.tradier.com/v1/markets/search"
tradier_access_token = os.getenv("TRADIER_ACCESS_TOKEN")
headers = {"Accept": "application/json", "Authorization" : f"Bearer {tradier_access_token}"}
```

```python
def get_matching_tickers(q):
    """
    Return the list of matching tickers for query term.
    """

    query_terms = q.split(" ")
    num_terms = len(query_terms)
    params = {"q": q, "lang": "en"}

    response = requests.get(tradier_lookup_url, headers=headers, params=params)
    response.raise_for_status()
    quotes = response.json()["securities"]

    # No matching results
    if quotes is None:
        if num_terms > 1:
            return get_matching_tickers(" ".join(query_terms[:num_terms-1]))
        else:
            return None

    quotes = quotes["security"]
    if isinstance(quotes, list):
        if len(quotes) > 0:
            return quotes[0]
        return None
    elif isinstance(quotes, dict):
        return quotes
    
    return None
```

```python

weapons_tickers = []
for name in weapons_df["Fund"].values:
    result = get_matching_tickers(name)
    if result is None:
        weapons_tickers.append('Not found')
        print(f"     !!{name}: Ticker not found")
    else:
        ticker = result["symbol"]
        weapons_tickers.append(ticker)
        print(f"{name}: {ticker}")

weapons_df["Ticker"] = weapons_tickers
```

```python
animal_tickers = []
for name in animal_df["Fund"].values:
    result = get_matching_tickers(name)
    if result is None:
        animal_tickers.append('Not found')
        print(f"     !!{name}: Ticker not found")
    else:
        ticker = result["symbol"]
        animal_tickers.append(ticker)
        print(f"{name}: {ticker}")

animal_df["Ticker"] = animal_tickers
```

```python
rights_tickers = []
for name in rights_df["Fund"].values:
    result = get_matching_tickers(name)
    if result is None:
        rights_tickers.append('Not found')
        print(f"     !!{name}: Ticker not found")
    else:
        ticker = result["symbol"]
        rights_tickers.append(ticker)
        print(f"{name}: {ticker}")

rights_df["Ticker"] = rights_tickers

```

```python
envir_tickers = []
for name in envir_df["Fund"].values:
    result = get_matching_tickers(name)
    if result is None:
        envir_tickers.append('Not Found')
        print(f"     !!{name}: Ticker not found")
    else:
        ticker = result["symbol"]
        envir_tickers.append(ticker)
        print(f"{name}: {ticker}")

envir_df["Ticker"] = envir_tickers
```

```python
weapons_df.drop(weapons_df.loc[weapons_df['Ticker']== 'Not found'].index, inplace=True)
envir_df.drop(envir_df.loc[envir_df['Ticker']== 'Not found'].index, inplace=True)
animal_df.drop(animal_df.loc[animal_df['Ticker']== 'Not found'].index, inplace=True)
rights_df.drop(rights_df.loc[rights_df['Ticker']== 'Not found'].index, inplace=True)
weapons_df=weapons_df.drop(columns='Defense/ Weapons')
envir_df.drop(columns='Environment')
animal_df.drop(columns='Animal Testing')
rights_df.drop(columns='Human Rights')
```
